# Linear regression

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import csv

csv_file = 'preprocessed_data/preprocessed_merged_apartments_population_2011-2022.csv'

apartments_df = pd.read_csv(csv_file)

apartments_df.head()
```

##### Sales price vs. population

```python
cities = apartments_df['City'].unique()
rooms_categories = ["", "one-room", "two-room", "three-room"]

for city in cities:
    for room_amount in range(1, 4):
        city_data = apartments_df[apartments_df['City'] == city]
        rooms = city_data[city_data["Rooms"]==room_amount]
    
        print(f"City: {city}")

        X = rooms['Total population'].values.reshape(-1, 1)
        y = rooms['Sales price per m2'].values
    
        model = LinearRegression().fit(X, y)
        slope = model.coef_[0]
        print(f"Slope for {rooms_categories[room_amount]} properties (Population Coefficient): {slope:.2f}")
    
        y_pred = np.zeros(len(y))
    
        for i in range(len(X)):
            X_train = np.concatenate((X[:i], X[i+1:]))
            X_test = X[i].reshape(-1, 1)
            y_train = np.concatenate((y[:i], y[i+1:]))
            model = LinearRegression().fit(X_train, y_train)
            y_pred[i] = model.predict(X_test)[0]
        

        print(f"Mean squared error for {rooms_categories[room_amount]} properties: {mean_squared_error(y, y_pred): .2f}")
        print(f"Coefficient of determination for {rooms_categories[room_amount]} properties: {r2_score(y, y_pred): .2f}")    

    sns.regplot(x="Total population", y="Sales price per m2", data=city_data[city_data["Rooms"]==1], ci=None, label="One room")
    sns.regplot(x="Total population", y="Sales price per m2", data=city_data[city_data["Rooms"]==2], ci=None, label="Two room")
    sns.regplot(x="Total population", y="Sales price per m2", data=city_data[city_data["Rooms"]==3], ci=None, label="Three room")
    plt.legend()
    plt.show()
```

##### Linear regression for population

```python
big_cities = ["Helsinki", "Espoo-Kauniainen", "Tampere", "Turku", "Jyväskylä", "Oulu"]
past_years = list(range(2011, 2023))
future_years = list(range(2023, 2033))

with open('preprocessed_data/prediction_data_year_price.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Year,City,Rooms,Sales price per m2"])

    for city in big_cities:
        city_data = apartments_df[apartments_df['City'] == city]
        X = city_data['Year'].unique().reshape(-1, 1)
        y = city_data['Total population'].unique()
    
        population_model = LinearRegression().fit(X, y)
        
        future_years = np.array(range(2023, 2033)).reshape(-1, 1)
        future_population = population_model.predict(future_years).reshape(-1, 1)
    
        for room_amount in range(1, 4):
            rooms = city_data[city_data["Rooms"]==room_amount]
    
            X = rooms['Total population'].values.reshape(-1, 1)
            y = rooms['Sales price per m2'].values
            
            prices_model = LinearRegression().fit(X, y)
            future_sales_prices = [int(x) for x in prices_model.predict(future_population)]
                    
            for index in range(len(past_years)):
                writer.writerow([past_years[index], city, room_amount, y[index]])
                
            for index in range(len(future_years)):
                writer.writerow([future_years[index][0], city, room_amount, future_sales_prices[index]])
```

##### Rent price vs. sales price

```python
for city in cities:
    city_data = apartments_df[apartments_df['City'] == city]
    one_room = city_data[city_data["Rooms"]==1]
    two_room = city_data[city_data["Rooms"]==2]
    three_room = city_data[city_data["Rooms"]==3]
        
    X = city_data['Rent per m2'].values.reshape(-1, 1)
    y = city_data['Sales price per m2'].values
 
    print(f"City: {city}")

    X = one_room['Rent per m2'].values.reshape(-1, 1)
    y = one_room['Sales price per m2'].values
    
    model = LinearRegression().fit(X, y)
    slope = model.coef_[0]
    
    print(f"Slope one room: {slope:.2f}")

    X = two_room['Rent per m2'].values.reshape(-1, 1)
    y = two_room['Sales price per m2'].values
    
    model = LinearRegression().fit(X, y)
    slope = model.coef_[0]
    
    print(f"Slope two room: {slope:.2f}")

    X = three_room['Rent per m2'].values.reshape(-1, 1)
    y = three_room['Sales price per m2'].values
    
    model = LinearRegression().fit(X, y)
    slope = model.coef_[0]
    
    print(f"Slope three room: {slope:.2f}")

    sns.regplot(x="Rent per m2", y="Sales price per m2", data=city_data[city_data["Rooms"]==1], ci=None, label="One room")
    sns.regplot(x="Rent per m2", y="Sales price per m2", data=city_data[city_data["Rooms"]==2], ci=None, label="Two room")
    sns.regplot(x="Rent per m2", y="Sales price per m2", data=city_data[city_data["Rooms"]==3], ci=None, label="Three room")
    plt.legend()
    plt.show()
```
